```python
num_ports = 15000

all_weights1 = np.zeros((num_ports,len(stocks1.columns)))
all_weights2 = np.zeros((num_ports,len(stocks2.columns)))
all_weights3 = np.zeros((num_ports,len(stocks3.columns)))
all_weights4 = np.zeros((num_ports,len(stocks4.columns)))
ret_arr1 = np.zeros(num_ports)
ret_arr2 = np.zeros(num_ports)
ret_arr3 = np.zeros(num_ports)
ret_arr4 = np.zeros(num_ports)
vol_arr1 = np.zeros(num_ports)
vol_arr2 = np.zeros(num_ports)
vol_arr3 = np.zeros(num_ports)
vol_arr4 = np.zeros(num_ports)
sharpe_arr1 = np.zeros(num_ports)
sharpe_arr2 = np.zeros(num_ports)
sharpe_arr3 = np.zeros(num_ports)
sharpe_arr4 = np.zeros(num_ports)


for ind in range(num_ports):

    weights1 = np.array(np.random.random(3))
    weights2 = np.array(np.random.random(3))
    weights3 = np.array(np.random.random(3))
    weights4 = np.array(np.random.random(3))

    weights1 = weights1 / np.sum(weights1)
    weights2 = weights2 / np.sum(weights2)
    weights3 = weights3 / np.sum(weights3)
    weights4 = weights4 / np.sum(weights4)

    all_weights1[ind,:] = weights1
    all_weights2[ind,:] = weights2
    all_weights3[ind,:] = weights3
    all_weights4[ind,:] = weights4

    ret_arr1[ind] = np.sum((log_ret1.mean() * weights1) *252)
    ret_arr2[ind] = np.sum((log_ret2.mean() * weights2) *252)
    ret_arr3[ind] = np.sum((log_ret3.mean() * weights3) *252)
    ret_arr4[ind] = np.sum((log_ret4.mean() * weights4) *252)

    vol_arr1[ind] = np.sqrt(np.dot(weights1.T, np.dot(log_ret1.cov() * 252, weights1)))
    vol_arr2[ind] = np.sqrt(np.dot(weights2.T, np.dot(log_ret2.cov() * 252, weights2)))
    vol_arr3[ind] = np.sqrt(np.dot(weights3.T, np.dot(log_ret3.cov() * 252, weights3)))
    vol_arr4[ind] = np.sqrt(np.dot(weights4.T, np.dot(log_ret4.cov() * 252, weights4)))



    sharpe_arr1[ind] = ret_arr1[ind]/vol_arr1[ind]
    sharpe_arr2[ind] = ret_arr2[ind]/vol_arr1[ind]
    sharpe_arr3[ind] = ret_arr3[ind]/vol_arr1[ind]
    sharpe_arr4[ind] = ret_arr4[ind]/vol_arr1[ind]
```

# Mathematical Optimization
 Use of optimization functions to find the ideal weights mathematically!

```python
def get_ret_vol_sr1(weights1):

    weights1 = np.array(weights1)

    ret1 = np.sum(log_ret1.mean() * weights1) * 252

    vol1 = np.sqrt(np.dot(weights1.T, np.dot(log_ret1.cov() * 252, weights1)))
    sr1 = ret1/vol1
    return np.array([ret1,vol1,sr1])
```

```python
def get_ret_vol_sr2(weights2):

    weights2 = np.array(weights2)

    ret2 = np.sum(log_ret2.mean() * weights2) * 252

    vol2 = np.sqrt(np.dot(weights2.T, np.dot(log_ret2.cov() * 252, weights2)))
    sr2 = ret2/vol2
    return np.array([ret2,vol2,sr2])
```

```python
def get_ret_vol_sr3(weights3):

    weights3 = np.array(weights3)

    ret3 = np.sum(log_ret3.mean() * weights3) * 252

    vol3 = np.sqrt(np.dot(weights3.T, np.dot(log_ret3.cov() * 252, weights3)))
    sr3 = ret3/vol3
    return np.array([ret3,vol3,sr3])
```

```python
def get_ret_vol_sr4(weights4):

    weights4 = np.array(weights4)

    ret4 = np.sum(log_ret4.mean() * weights4) * 252

    vol4 = np.sqrt(np.dot(weights4.T, np.dot(log_ret4.cov() * 252, weights4)))
    sr4 = ret4/vol4
    return np.array([ret4,vol4,sr4])
```

```python
from scipy.optimize import minimize
```

Optimization works as a minimization function, since we actually want to maximize the Sharpe Ratio, we will need to turn it negative so we can minimize the negative sharpe

```python
def neg_sharpe1(weights1):
    return  get_ret_vol_sr1(weights1)[2] * -1
```

```python
def neg_sharpe2(weights2):
    return  get_ret_vol_sr2(weights2)[2] * -1
```

Efficient Frontier:

The efficient frontier is the set of optimal portfolios that offers the highest expected return for a defined level of risk or the lowest risk for a given level of expected return. Portfolios that lie below the efficient frontier are sub-optimal, because they do not provide enough return for the level of risk. Portfolios that cluster to the right of the efficient frontier are also sub-optimal, because they have a higher level of risk for the defined rate of return.

Efficient Frontier http://www.investopedia.com/terms/e/efficientfrontier
